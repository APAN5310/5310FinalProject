```python
stmt = """
CREATE TABLE Railroad (
 Reporting_Railroad_Code  varchar(100) , 
 Reporting_Railroad_Name  varchar(800) , 
 PRIMARY KEY (Reporting_Railroad_Code)
) ; 

CREATE TABLE State (
 State_Code  numeric(3,1) , 
 State_Abbreviation varchar(5) , 
 State_Name   varchar(100) , 
 PRIMARY KEY (State_Code)
) ; 



CREATE TABLE Weather_Condition (
 Weather_Condition_Code  numeric(2,1) , 
 Weather_Condition   varchar(100) , 
 PRIMARY KEY (Weather_Condition_Code)
) ; 


CREATE TABLE Train_Direction (
 Train_Direction_Code varchar(5) , 
 Train_Direction   varchar(50) , 
 PRIMARY KEY (Train_Direction_Code)
) ; 

CREATE TABLE Equipment (
 Equipment_Type_Code  varchar(50) , 
 Equipment_Type   varchar(300) , 
 PRIMARY KEY (Equipment_Type_Code)
) ;



CREATE TABLE primary_accident_cause_code (
  Primary_Accident_Cause_Code VARCHAR(20) PRIMARY KEY,
  Primary_Accident_Cause VARCHAR(500) NOT NULL
);

CREATE TABLE contributing_accident_cause_code (
  Contributing_Accident_Cause_Code VARCHAR(20) PRIMARY KEY,
  Contributing_Accident_Cause VARCHAR(200) NOT NULL
);

CREATE TABLE accident_cause_code (
  Accident_Cause_Code VARCHAR(20) PRIMARY KEY,
  Accident_Cause VARCHAR(200) NOT NULL
);

CREATE TABLE Visibility (
  Visibility_Code numeric(2,1) PRIMARY KEY,
  Visibility VARCHAR(50) NOT NULL
);




CREATE TABLE Accident ( 
 Accident_Number VARCHAR(100)PRIMARY KEY ,
 Accident_Date date,
 Reporting_Railroad_Code  varchar(100),
 Visibility_Code numeric(2,1),
 Weather_Condition_Code numeric(2,1), 
 Train_Direction_Code varchar(5),
 Equipment_Type_Code CHAR(50),  
 Primary_Accident_Cause_Code VARCHAR(20), 
 Contributing_Accident_Cause_Code VARCHAR(20),
 Accident_Cause_Code VARCHAR(20),
 FOREIGN KEY (Reporting_Railroad_Code) REFERENCES Railroad (Reporting_Railroad_Code),
 FOREIGN KEY (Visibility_Code) REFERENCES Visibility (Visibility_Code),
 FOREIGN KEY (Weather_Condition_Code) REFERENCES Weather_Condition (Weather_Condition_Code),
 FOREIGN KEY (Train_Direction_Code) REFERENCES Train_Direction (Train_Direction_Code),
 FOREIGN KEY (Equipment_Type_Code) REFERENCES Equipment (Equipment_Type_Code),
 FOREIGN KEY (Primary_Accident_Cause_Code) REFERENCES primary_accident_Cause_code (Primary_Accident_Cause_Code),
 FOREIGN KEY (Contributing_Accident_Cause_Code) REFERENCES contributing_accident_cause_code (Contributing_Accident_Cause_Code),
 FOREIGN KEY (Accident_Cause_Code) REFERENCES accident_cause_code (Accident_Cause_Code)
);


CREATE TABLE Injured (
    Injured_Id     int,
    Accident_Number VARCHAR(100),
    Railroad_Employees_Killed   numeric(10,3),
    Railroad_Employees_Injured  numeric(10,3),
    Passengers_Killed  numeric(10,3), 
    Passengers_Injured  numeric(10,3), 
    Others_Killed numeric(10,3), 
    Others_Injured numeric(10,3), 
    Persons_Killed_For_Reporting_Railroad numeric(10,3), 
    Persons_Injured_For_Reporting_Railroad numeric(10,3), 
    Total_Persons_Killed numeric(10,3), 
    Total_Persons_Injured numeric(10,3), 
    Total_Killed_Form_54 numeric(10,3), 
    Total_Injured_Form_54 numeric(10,3), 
    Primary Key (Injured_Id),
    FOREIGN KEY (Accident_Number) REFERENCES Accident (Accident_Number)
);

CREATE TABLE Accident_Location (
 Location_id int PRIMARY KEY,
 Accident_Number VARCHAR(100),
 Latitude numeric(10,7),
 Longitude numeric(10,7),
 Narrative varchar(2500),
 State_Code numeric(3,1),
 FOREIGN KEY (State_Code) REFERENCES State (State_Code),
 FOREIGN KEY (Accident_Number) REFERENCES Accident (Accident_Number)
);


CREATE TABLE damage_costs (
  Damage_Costs_id INT PRIMARY KEY,
  Accident_Number VARCHAR(100),
  Equipment_Damage_Costs INT,
  Track_Damage_Costs INT,
  Total_Damage_Costs INT,
  FOREIGN KEY (Accident_Number) REFERENCES Accident (Accident_Number)
);




CREATE TABLE Train_speed (
  Train_speed_id INT PRIMARY KEY,
  Accident_Number VARCHAR(100),
  Train_Speed numeric(5,2),
  Recorded_Estimated_Speed varchar(100),
  Maximum_Speed INT,
  FOREIGN KEY (Accident_Number) REFERENCES Accident (Accident_Number)
);




CREATE TABLE Train_load (
    Load_id int PRIMARY KEY,
	Accident_Number VARCHAR(100),
    Loaded_Freight_Cars numeric(10,2) NOT NULL,
    Gross_Tonnage integer NOT NULL,
	FOREIGN KEY (Accident_Number) REFERENCES Accident (Accident_Number)
);


"""
connection.execute(stmt)
```
